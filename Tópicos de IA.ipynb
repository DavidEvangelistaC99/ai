```python
# Diseño de Perceptrón multicapa
# Pasos para el diseño y obtención de una red neuronal

# 1. Cargar datos y tenerlos listos para modelar.
# 2. Diseño de perceptrón multicapa (Keras).
# 3. Compilar diseño de perceptrón multicapa (Keras).
# 4. Ajustamos perceptrón al conjunto de datos.
# 5. Evaluamos el modelo obtenido.
# 6. Juntar/Empaquetar la obtención de nuestro perceptrón en una función/clase.
# 7. Creación de predicciones.
```

```python
# 1. Cargar datos y tenerlos listos para modelar.
# Importación de módulos y datos

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importación del modelo secuencial de Keras
# Este tipo de modelo es una pila lineal de capas donde cada capa tiene exactamente una entrada y una salida.
from keras.models import Sequential

# Dense: Importación de capas que conecta las neuronas de la capa actual con todas de la siguiente capa.
# Las neuronas se comunican con todas las demás (capa densa)
# Input: Especifica la forma de entrada del modelo
from keras.layers import Dense, Input

# Datos
# https://github.com/robintux/Datasets4StackOverFlowQuestions/blob/master/pima-indians-diabetes.csv
dataset = pd.read_csv('pima-indians-diabetes.csv', header = None)

# Definimos las variables independientes y la dependiente
X = dataset.loc[:,0:7]
y = dataset.loc[:,8]
```

```python
# Analizamos la distribución de los valores de la variable dependiente
y.value_counts().plot(kind = 'bar')
plt.show()
print(y.value_counts())
```

```python
# 2. Diseño de perceptrón multicapa (Keras).
# Diseñamos nuestra arquitectura de red neuronal

# Instancia de un objeto de la clase Sequencial
rnn1 = Sequential()

# Definimos la capa de entrada
rnn1.add(Input(shape = (X.shape[1], ), name = 'Capa de entrada'))

# Agregamos el total de capas ocultas
rnn1.add(Dense(units = 4, activation = 'relu'))
rnn1.add(Dense(units = 4, activation = 'elu'))
rnn1.add(Dense(units = 2, activation = 'tanh'))
rnn1.add(Dense(units = 5, activation = 'selu'))

# Definimos la capa de salida
rnn1.add(Dense(units = 1, activation = 'sigmoid'))

# Compilamos la arquitectura
# loss: binary_crossentropy: Función de pérdida utilizada en clasificación binaria
# optimizer: adam: Ajuste de pesos en función del valor entregado por la función de pérdida
# metrics: Métricas de rendimiento del modelo durante el entrenamiento y validación
# accuracy: Mide el porcentaje de predicciones correctas
rnn1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Ajustamos la arquitectura de la red neuronal (nuestro conjunto de datos)
# Entrenamiento del modelo con los datos
rnn1.fit(X, y, epochs = 200)

# Calculamos los pronosticos por no tener un metodo de testeo
# Método para realizar predicciones (mismos datos)
predic_y = rnn1.predict(X)
```

```python
# Visualizacion de los pronosticos para y : predic_y
plt.scatter(x = range(len(predic_y)), y = predic_y)

# Establecer el límite del eje y entre 0 y 1
plt.ylim(0, 1)
plt.show()
```

```python
# Convertimos las probabilidades en etiquetas binarias (umbral 0.5)
pred_y = np.where(predic_y > 0.5, 1, 0)

# Indicador de calidad del modelo
from sklearn import metrics
metrics.accuracy_score(y, pred_y)
```
